```python
def scrape_medication_info(medication_name):
    url = f"https://www.rxlist.com/{medication_name}-drug.htm"
    response = requests.get(url)
    
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, "html.parser")
        
        # Find the section containing the medication information
        medication_info_section = soup.find("section", class_="story")
        
        if medication_info_section:
            # Extract all text within the section
            medication_info_text = medication_info_section.get_text(separator="\n")
            
            # Print or do whatever you want with the extracted text
            print("Medication Information:")
            print(medication_info_text)
        
        else:
            print("Medication information section not found on the webpage.")
        
    else:
        print("Failed to retrieve the webpage.")

# Example usage
scrape_medication_info("aspirin")
```
